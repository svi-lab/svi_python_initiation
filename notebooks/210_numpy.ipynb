## Indexing with boolean arrays (masks)

```python
a = np.arange(20); a
```

```python
mask1 = a==2;mask1
```

```python
mask = a % 2 == 0; mask
```

```python
a[mask] = -1; print(a)
```

```python
random_nb = np.random.random(10)
print(random_nb)
```

```python
mask = random_nb > 0.5
random_nb[mask] = -2
print(random_nb)
```
